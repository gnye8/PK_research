```python
import arnie
from arnie.utils import *
import pandas as pd

def get_sliding_windows(full_struct, step, window):
    coords = list(range(0,len(full_struct)-window+1,step))
    windows = []
    for i in coords:
        new_window = full_struct[i:i+window]
        windows.append(new_window)
    return windows, coords

def get_pseudoknots(structure, step, window, output_folder):
    struct = structure.replace('-', '')
    windows,coords = get_sliding_windows(struct, step=step, window=window)
    PK_list = []
    for struct,coord in zip(windows,coords):
        if is_PK(struct):
            PK_list.append([coord, coord+window, struct])
    df = pd.DataFrame(PK_list,columns=["start","end", "struct"])
    df.to_csv(f'{output_folder}/pk_predictor_output.csv')
```
