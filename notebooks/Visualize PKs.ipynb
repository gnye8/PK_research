```python
import pandas as pd
import seaborn as sns

knotty = pd.read_csv('/home/gnye8/Desktop/PK_research/pipeline_results/knotty/knotty_analysis_scores.csv')

pknots = pd.read_csv('/home/gnye8/Desktop/PK_research/pipeline_results/pknots/pknots_analysis_scores.csv')
```

```python
knotty.sort_values('pk_bp_shape_score', ascending=False)
```

```python
pknots.sort_values('pk_bp_shape_score', ascending=False)
```

```python
# plotting: shape vs. pk bp shape score
# shape score vs. rank
# shape score vs. consensus

g = sns.jointplot(x="shape_score", y="pk_bp_shape_score", data=knotty, kind="reg")
g = sns.jointplot(x='shape_score', y='pk_bp_shape_score', data=pknots, kind='reg')

g = sns.jointplot(x="shape_score", y="pk_shape_score", data=knotty, kind="reg")
g = sns.jointplot(x="shape_score", y="pk_shape_score", data=pknots, kind="reg")
```

```python
# shape score vs. rank

g = sns.jointplot(x='pk_bp_shape_score', y='rank', data=knotty, kind='reg')
g = sns.jointplot(x='pk_bp_shape_score', y='rank', data=pknots, kind='reg')

g = sns.jointplot(x='shape_score', y='rank', data=knotty, kind='reg')
g = sns.jointplot(x='shape_score', y='rank', data=pknots, kind='reg')
```

```python
# shape score vs. consensus

g = sns.jointplot(x='pk_bp_shape_score', y='bp_pk_consensus_score', data=knotty, kind='reg')
g = sns.jointplot(x='pk_bp_shape_score', y='bp_pk_consensus_score', data=pknots, kind='reg')

g = sns.jointplot(x='shape_score', y='consensus_score', data=knotty, kind='reg')
g = sns.jointplot(x='shape_score', y='consensus_score', data=pknots, kind='reg')
```

```python
# rank vs. consensus

g = sns.jointplot(x="rank", y="consensus_score", data=knotty, kind="reg")
g = sns.jointplot(x="rank", y="consensus_score", data=pknots, kind="reg")

g = sns.jointplot(x="rank", y="bp_pk_consensus_score", data=knotty, kind="reg")
g = sns.jointplot(x="rank", y="bp_pk_consensus_score", data=pknots, kind="reg")
```
